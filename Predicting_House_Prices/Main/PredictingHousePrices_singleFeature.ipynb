# Load House Sales Data

```python
sales = graphlab.SFrame('home_data.sframe/')
```

```python
sales
```

```python
sales.print_rows(3,7)
```

```python
sales.print_rows(21000,5)
```

# Exploring data

```python
graphlab.canvas.set_target('ipynb')
sales.show(view = 'Scatter Plot',x='sqft_living',y='price')
```

# Creating a simple regression

## Split the data into training and test datasets

```python
train_data,test_data = sales.random_split(0.8,seed=0)
```

## Build the regression model

```python
lin_regr_prediction = graphlab.linear_regression.create(train_data,target='price',features=['sqft_living'])
```

# Evaluate Our Model

```python
print (test_data['price'].mean())
```

```python
print (lin_regr_prediction.evaluate(test_data))
```

# What are our predictions look like

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_data['sqft_living'],test_data['price'],'.',
        test_data['sqft_living'],lin_regr_prediction.predict(test_data),'-')
```

## Get Coefficients (w0,w1)

```python
lin_regr_prediction.get('coefficients')
```
