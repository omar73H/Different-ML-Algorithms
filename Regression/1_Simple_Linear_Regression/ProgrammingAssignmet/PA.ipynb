```python
train_data , test_data = house_prices.random_split(.8,seed=0)
```

```python
def simple_linear_regression(input_feature , target):
    n = input_feature.size()
    sum_x = input_feature.sum()
    sum_y = target.sum()
    sum_xy = (input_feature * target).sum()
    sum_xx = (input_feature * input_feature).sum()
    
    slope = (n * sum_xy - sum_x * sum_y)  / (n * sum_xx - sum_x * sum_x)
    intercept = sum_y/n - slope * (sum_x/n)
    
    return (intercept,slope)
```

```python
input_feature = train_data['sqft_living']
target = train_data['price']

(sqft_intercept, sqft_slope) = simple_linear_regression(input_feature , target)
print sqft_intercept
print sqft_slop
```

```python
def get_regression_predictions(input_feature, intercept, slope):
    predicted_target = input_feature.apply(lambda x: intercept + slope*x)
    
    return predicted_target
```

```python
y_2650_ft = sqft_intercept + sqft_slope*2650 ;
```

```python
def get_residual_sum_of_squares(input_feature, real_output, intercept, slope):
    # RSS = SUM ( Yi - (intercept + slope * Xi) )^2
    # real_output --> Yi
    predicted_output = get_regression_predictions(input_feature, intercept, slope)
    
    values = (real_output - predicted_output).apply(lambda x : x**2)
    
    RSS = values.sum()
    
    return RSS
```

```python
train_RSS = get_residual_sum_of_squares(input_feature, target, sqft_intercept, sqft_slope)
```

```python
def inverse_regression_predictions(output, intercept, slope):
    estimated_input = output.apply(lambda x: (x-intercept)/slope)
    return estimated_input
```

```python
x_800000_dollar = (800000-sqft_intercept)/sqft_slope
```
